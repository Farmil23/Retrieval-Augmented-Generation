```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_community.document_loaders import PyMuPDFLoader
from langchain_core.documents.base import Document
from typing import List
```

```python
class SmartPreprocessesChunk:
    def __init__(self, chunk_size=2000, chunk_overlap= 200):
        self.chunk_size = chunk_size
        self.chunk_overlap = chunk_overlap
        self.text_splitter = RecursiveCharacterTextSplitter(
            chunk_size=chunk_size,
            chunk_overlap= chunk_overlap
        )
        
    def Preprocessed_chunk(self, pdf_path:str) -> List[Document]:
        loader = PyMuPDFLoader(pdf_path)
        pages = loader.load()
        
        for page_num , content in enumerate(pages):
            
            cleaned_text = _cleaned_text(content.page_content)
            
            documents = []
            if cleaned_text.strip() < 50:
                content
                
            chunks = self.text_splitter.split_text(
                texts = [cleaned_text],
                metadata={
                    **content.metadata,
                    "page_number" : page_num,
                    "sources" : pdf_path,
                    "MethodChunks" : SmartPreprocessesChunk
                }
            )
            
            documents.extend(chunks)
            
        return documents
    
    def _cleaned_text(self, text):
        cleaned_text = ' '.join(text.split())
        return cleaned_text
    

            
```
